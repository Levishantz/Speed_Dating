Opposites Attracts: Do Couples with Different Careers Match Up?

The data set provided in "Speed Dating" from Kaggle included each participant's careers coded into separate categories.  

career_c: career coded 
1= Lawyer 
2= Academic/Research 
3= Psychologist 
4= Doctor/Medicine 
5=Engineer 
6= Creative Arts/Entertainment 
7=Banking/Consulting/Finance/Marketing/Business/CEO/Entrepreneur/Admin 
8= Real Estate 
9= International/Humanitarian Affairs 
10= Undecided 
11=Social Work
12=Speech Pathology
13=Politics
14=Pro sports/Athletics
15=Other
16=Journalism
17=Architecture

In an attempt to prove our null hypothesis, I compared the career categories of the participants and their partners to analyze if similarlity in careers can determine a match.

```python
# %matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
data_to_load = "Resources/speed_dating_data.csv"

# Read Data

data = pd.read_csv(data_to_load, encoding="ISO-8859-1")
data.head()
```

```python
sd_csv = "Resources/speed_dating_data.csv"
ind_df = pd.read_csv(sd_csv, encoding="ISO-8859-1")
par_df = pd.read_csv(sd_csv, encoding="ISO-8859-1")

clean_ind = ind_df.copy()
clean_ind = clean_ind[["iid","pid","match","career_c","match_es"]]
clean_ind.head()

clean_par = par_df.copy()
clean_par = clean_par[["iid","pid","match","career_c","match_es"]]
clean_par.head()

combo_df = pd.merge(clean_ind,clean_par, left_on = 'iid',right_on='pid')
combo_df = combo_df[combo_df['pid_x'] == combo_df['iid_y']]
combo_df

found = []
found_sets = []
for x, y in zip(combo_df['pid_x'], combo_df['iid_x']):
    if set([x, y]) in found_sets:
        found.append(False)
    else:
        found.append(True)
        found_sets.append(set([x, y]))
combo_df = combo_df[found]

combo_df.head()
```

```python
# Drops No Matches
data_drop2 = combo_df.loc[combo_df["match_x"] == 1, :]
data_drop2.head()
```

```python
# Counts the Number of Matches between the career of the partner and the participant
df_new = data_drop2.groupby(['career_c_x', 'career_c_y']).count()
df_new.head()
```

```python
# Resets Index
df_new = df_new.reset_index()
df_new.columns
```

```python
# Creates New DF
clean_data = pd.DataFrame({"Career": df_new['career_c_x'],"Partner Career":df_new['career_c_y'],"Matches":df_new['match_x']})
clean_data.head()
```

```python
# Creates New DF with Percentage of Total Matches 
clean_data['Percentage'] = (clean_data['Matches']/clean_data['Matches'].sum())*100
clean_data.max()
```

```python
#Creates Heatmap of Percentages of the Career Categories between the pairings
# import seaborn as sns
%matplotlib inline

import pandas as pd
import seaborn as sns
# import matplotlib as plt
import matplotlib.pyplot as plt

x = clean_data['Career']
y = clean_data['Partner Career']
z = clean_data['Percentage']

data = pd.DataFrame(data={'Career':x, 'Partner Career':y, 'z':z})
data = data.pivot(index='Career', columns='Partner Career', values='z')
heatmap1 = sns.heatmap(data, cmap="BuPu").set_title("Percentages of Matches Between Career Categories")
plt.gcf().subplots_adjust(bottom=0.20)
heatmap1_fig = heatmap1.get_figure()

plt.savefig("Percentages of Matches Between Career Categories.png")
plt.savefig("Percentages of Matches Between Career Categories.png",
transparent=False)
```

10%+ matches between 2 & 2 career categories
10%+ matches between 7 & 2 career categories
Therefore the two highest number of matches occurred between simliar careers and dissimilar careers
Majority of participants in business/finance field
Insignificant findings of matches in other career categories

Unable to conclude that matches between similar career paths occurred between the participants and their parteners at this time.

Next, I will compare the career catergories among parteners who did not match.

```python
sd_csv2 = "Speed Dating Data.csv"
ind_df2 = pd.read_csv(sd_csv2, encoding="ISO-8859-1")
par_df2 = pd.read_csv(sd_csv2, encoding="ISO-8859-1")

clean_ind2 = ind_df2.copy()
clean_ind2 = clean_ind2[["iid","pid","match","career_c","match_es"]]
clean_ind2.head()

clean_par2 = par_df2.copy()
clean_par2 = clean_par2[["iid","pid","match","career_c","match_es"]]
clean_par2.head()

combo_df2 = pd.merge(clean_ind2,clean_par2, left_on = 'iid',right_on='pid')
combo_df2 = combo_df2[combo_df2['pid_x'] == combo_df2['iid_y']]
combo_df2

found2 = []
found_sets2 = []
for x, y in zip(combo_df2['pid_x'], combo_df2['iid_x']):
    if set([x, y]) in found_sets2:
        found2.append(False)
    else:
        found2.append(True)
        found_sets2.append(set([x, y]))
combo_df2 = combo_df2[found]

combo_df2.head()
```

```python
# Drops Matches
Nondata_drop2 = combo_df2.loc[combo_df2["match_x"] == 0, :]
Nondata_drop2.head()
```

```python
# Counts the Number of Non-Matches between the career of the partner and the participant
NONdf_new = Nondata_drop2.groupby(['career_c_x', 'career_c_y']).count()
NONdf_new.head()
```

```python
# Resets Index
NONdf_new = NONdf_new.reset_index()
NONdf_new.columns
```

```python
# Creates New DF
NONclean_data = pd.DataFrame({"Career": NONdf_new['career_c_x'],"Partner Career":NONdf_new['career_c_y'],"Non Matches":NONdf_new['match_x']})
NONclean_data.head()
```

```python
# Creates New DF with Percentage of Total Matches 
NONclean_data['Percentage'] = (NONclean_data['Non Matches']/NONclean_data['Non Matches'].sum())*100
NONclean_data.head()
```

```python
#Creates Heatmap of Percentages of the Career Categories between the pairings
# import seaborn as sns
%matplotlib inline

import pandas as pd
import seaborn as sns
# import matplotlib as plt
import matplotlib.pyplot as plt

x = NONclean_data['Career']
y = NONclean_data['Partner Career']
z = NONclean_data['Percentage']

data = pd.DataFrame(data={'Career':x, 'Partner Career':y, 'z':z})
data = data.pivot(index='Career', columns='Partner Career', values='z')
heatmap2 = sns.heatmap(data, cmap="BuPu").set_title("Percentages of Non-Matches Between Career Categories")
plt.gcf().subplots_adjust(bottom=0.20)
heatmap2_fig = heatmap2.get_figure()

plt.savefig("Percentages of Non-Matches Between Career Categories.png")
plt.savefig("Percentages of Non-Matches Between Career Categories.png",
transparent=False)
```

10%+ non matches between 2 & 2 career categories
10%+ non matches between 7 & 2 career categories
Therefore the two highest number of non matches occurred between simliar careers and dissimilar careers

Inconclusion, opposites do not attract in comparing careers as a variable.  Both matches and non matches showed similiar paterns.
